```python
import pandas as pd
```

```python
# reading the csv file
df_exams = pd.read_csv("StudentsPerformance.csv")
df_exams
```

# Rename Columns

```python
# rename column and overwrite the values of the dataframe
df_exams = df_exams.rename(columns = {"gender":"Gender"})
```

```python
# rename multiple columns and updating the dataframe with the inplace parameter
df_exams.rename(columns = {"math score" : "MS", "writing score":"WS", "reading score":"RS"}, inplace = True)
```

```python
# show dataframe
df_exams
```

# Rename Index

```python
# rename index 0, 1 and 2 and update the dataframe
df_exams.rename(index = {0:"A", 1:"B", 2:"C"}, inplace = True)
```
