## Adding a new column with an array

The dataframe has 1000 rows, so we need to create an array of 1000 numbers

```python
# importing numpy
import numpy as np
```

```python
# creating a range of 1000 numbers
language_score = np.arange(0, 1000, 1)
```

```python
# length of the array
len(language_score)
```

```python
df_exams["language score"] = language_score
```

```python
# showing the first 10 rows in the new dataframe
df_exams.head(10)
```

```python
# create random integer numbers between 1 and 100
int_language_score = np.random.randint(1, 100, size = 1000)
```

```python
# showing the maximum and minimum integer number
print(min(int_language_score))
print(max(int_language_score))
```

```python
# adding a new column to the dataframe with an array
df_exams["language score"] = int_language_score
```
